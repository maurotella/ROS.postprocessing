```python
import zlib, pandas as pd, sqlite3, zstd, pickle, base64
path = '/home/aislab/Documents/Tellaroli/test_ws/src/tesi.triennale.tellaroli/data/data_test.db'
conn = sqlite3.connect(path, check_same_thread=False)
```

```python
def object_to_dict(Obj):
    if '__slots__' not in dir(Obj): return Obj
    return {key:object_to_dict(Obj.__getattribute__(key)) for key in Obj.__slots__}
```

```python
data = pd.read_sql('SELECT * FROM FinalMaps WHERE execution=520', conn)
data
```

```python
cur = conn.cursor()
for i in range(len(data)):
    D = data.iloc[i]
    ex = D['execution']
    robot = D['robot']
    time = D['time']
    map = D['map']
    encoded = base64.b64encode(zlib.compress(pickle.dumps(object_to_dict(pickle.loads(zlib.decompress(map)))))).decode('ascii')
    sql = f"UPDATE FinalMaps SET map='{encoded}' WHERE execution={ex} AND robot={robot} AND time={time}"
    print(ex, robot, time, encoded)
    print(sql)
    cur.execute(
        sql
    )
conn.commit()
```
